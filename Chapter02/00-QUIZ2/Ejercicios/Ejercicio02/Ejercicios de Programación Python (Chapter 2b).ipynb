### Reto 2: Validador de Contraseñas
**Enunciado:** Crea una función que reciba un string y compruebe si es una contraseña válida. Debe tener más de 8 caracteres, al menos una letra mayúscula y al menos un número. Devuelve `True` o `False`.

**Pistas:** Usa `len()`, y combinaciones de `any()` con comprensiones de lista y los métodos `.isupper()` y `.isdigit()` sobre los caracteres.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 2
def validar_password(pwd):
    if len(pwd) <= 8:
        return False
    tiene_mayus = any([c.isupper() for c in pwd])
    tiene_num = any([c.isdigit() for c in pwd])
    return tiene_mayus and tiene_num

# Prueba:
password = input("Introduce una contraseña: ")
if validar_password(password):
    print("Contraseña válida")
else:
    print("Contraseña no válida")
```

### Reto 3: El Aplanador de Matrices
**Enunciado:** Tienes una matriz (lista de listas) como `[[1, 2], [3, 4], [5, 6]]`. Crea una función que la convierta en una lista plana `[1, 2, 3, 4, 5, 6]` usando solo una línea de código dentro de la función.

**Pistas:** Utiliza una comprensión de listas con dos bucles `for` (uno para la fila y otro para el elemento) en la misma línea.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 3
def aplanar(matriz):
    return [elemento for fila in matriz for elemento in fila]

# Prueba:
matriz = [[1, 2], [3, 4], [5, 6]]
print("Matriz original:", matriz)
print("Matriz aplanada:", aplanar(matriz))  
```

### Reto 4: Agrupación Categórica
**Enunciado:** Crea una función que reciba una lista de palabras mezcladas. Debe limpiar los espacios extra, pasarlas a minúsculas y agruparlas en un diccionario donde la clave sea la longitud de la palabra. Descarta las palabras que contengan números.

**Pistas:** Usa `defaultdict(list)` de la librería `collections`. Filtra validando con los métodos `.strip()`, `.lower()` y `.isalpha()`.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 4
from collections import defaultdict

def agrupar_palabras(lista_palabras):
    agrupadas = defaultdict(list)
    for palabra in lista_palabras:
        limpia = palabra.strip().lower()
        if limpia.isalpha():
            agrupadas[len(limpia)].append(limpia)
    return dict(agrupadas)

# Prueba:
palabras = ["  Hola ", "mundo", "Python3", "programación", "  ", "123", "prueba"]
print("Palabras originales:", palabras)
print("Palabras agrupadas por longitud:", agrupar_palabras(palabras))
```

### Reto 5: El Cajero Automático
**Enunciado:** Crea un bucle que pida por teclado una cantidad a retirar (asumiendo un saldo inicial de 1000). Captura el error si el usuario teclea letras. Si la cantidad supera el saldo, muestra un aviso. Siempre, haya error o no, debe imprimir "Operación de cajero finalizada" al terminar el intento.

**Pistas:** Usa `while True`, `input()`, un bloque `try/except ValueError` para las letras, y un bloque `finally`. Usa `break` cuando el retiro sea exitoso para salir del bucle.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 5
saldo = 1000
while True:
    try:
        cantidad = float(input("Cantidad a retirar: "))
        if cantidad > saldo:
            print("Saldo insuficiente.")
        else:
            saldo -= cantidad
            print(f"Retiro exitoso. Saldo restante: {saldo}")
            break
    except ValueError:
        print("Error: Introduce un número válido.")
    finally:
        print("Operación de cajero finalizada.\n")
```

### Reto 6: Detector de Anagramas
**Enunciado:** Crea una función que reciba dos palabras y determine si son anagramas (tienen exactamente las mismas letras). Ignora los espacios en blanco y las mayúsculas.

**Pistas:** Usa `.replace(" ", "")`, `.lower()` y compara ambas palabras pasándolas previamente por la función `sorted()`.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 6
def son_anagramas(p1, p2):
    p1_limpia = p1.replace(" ", "").lower()
    p2_limpia = p2.replace(" ", "").lower()
    return sorted(p1_limpia) == sorted(p2_limpia)

# Prueba:
palabra1 = input("Introduce la primera palabra: ")
palabra2 = input("Introduce la segunda palabra: ")
if son_anagramas(palabra1, palabra2):
    print(f"'{palabra1}' y '{palabra2}' son anagramas.")
else:
    print(f"'{palabra1}' y '{palabra2}' no son anagramas.")
```

### Reto 7: Fusión de Inventarios
**Enunciado:** Crea una función que reciba dos diccionarios de inventario (ej. `{"manzanas": 5}` y `{"manzanas": 3, "peras": 2}`). Debe devolver un nuevo diccionario combinando ambos, sumando las cantidades numéricas de las claves que coincidan.

**Pistas:** Puedes iterar sobre las claves y valores del segundo diccionario con `.items()`. Usa `.get(clave, 0)` sobre el primer diccionario para sumar valores sin provocar errores.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 7
def fusionar_inventarios(inv1, inv2):
    fusion = dict(inv1)  # Copiamos el primero
    for k, v in inv2.items():
        fusion[k] = fusion.get(k, 0) + v
    return fusion

# Prueba:
inventario1 = {"manzanas": 10, "naranjas": 5}
inventario2 = {"manzanas": 3, "peras": 7}
print("Inventario 1:", inventario1)
print("Inventario 2:", inventario2)
print("Inventario fusionado:", fusionar_inventarios(inventario1, inventario2))
```

### Reto 8: Buscador de Intrusos
**Enunciado:** Crea una función que reciba una palabra. Usando conjuntos (`set`), averigua y devuelve un nuevo conjunto solo con las letras de esa palabra que **no** sean vocales.

**Pistas:** Crea un conjunto con las vocales `{"a", "e", "i", "o", "u"}`. Convierte la palabra a minúsculas y luego a conjunto. Resta el conjunto de vocales al conjunto de la palabra.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 8
def consonantes_usadas(palabra):
    vocales = {"a", "e", "i", "o", "u"}
    letras_palabra = set(palabra.lower())
    letras_palabra.discard(" ")  # Por si tiene espacios
    return letras_palabra - vocales

# Prueba:
palabra = input("Introduce una palabra: ")
print("Palabra original:", palabra)
print("Consonantes usadas:", consonantes_usadas(palabra))
```

### Reto 9: Transposición de Tablero
**Enunciado:** Dada una matriz regular de 3x3 como `[[1, 2, 3], [4, 5, 6], [7, 8, 9]]`, crea una función que devuelva la matriz transpuesta (las filas se convierten en columnas y viceversa) usando comprensiones de lista.

**Pistas:** Necesitarás una comprensión de lista anidada iterando sobre los índices de las columnas (`range(len(matriz[0]))`).

```python
# Escribe tu código aquí
```

```python
# Solución Reto 9
def transponer(matriz):
    return [[fila[i] for fila in matriz] for i in range(len(matriz[0]))]

# Prueba:
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("Matriz original:")
for fila in matriz:
    print(fila)
print("Matriz transpuesta:")
for fila in transponer(matriz):
    print(fila)
```

### Reto 11: El Censor de Chat
**Enunciado:** Crea una función que reciba una frase y una lista de "palabras prohibidas". La función debe devolver la frase sustituyendo las palabras prohibidas por asteriscos (tantos como letras tenga la palabra).

**Pistas:** Itera sobre la lista de palabras prohibidas y usa el método `.replace()` en la frase. Para crear los asteriscos correctos, multiplica el string `"*"` por la longitud de la palabra con `len()`.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 11
def censurar(frase, prohibidas):
    for palabra in prohibidas:
        frase = frase.replace(palabra, "*" * len(palabra))
    return frase

# Prueba:
frase = input("Introduce una frase: ")
prohibidas = input("Introduce palabras prohibidas (separadas por comas): ").split(",")
prohibidas = [p.strip() for p in prohibidas]  # Limpiar espacios
print("Frase original:", frase)
print("Frase censurada:", censurar(frase, prohibidas))
```

### Reto 12: De CSV a Diccionario
**Enunciado:** Tienes una lista de cadenas que simula un archivo CSV: `["nombre,edad", "Ana,25", "Luis,30"]`. Crea una función que extraiga las cabeceras en la primera posición y cree un diccionario para cada registro, devolviendo una lista de diccionarios final.

**Pistas:** Extrae el primer elemento con `pop(0)` o segmentación y divídelo con `.split(',')`. Luego, itera el resto, divídelos también y usa `zip()` junto a `dict()` con las cabeceras.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 12
def procesar_csv(filas):
    cabeceras = filas.pop(0).split(",")
    resultado = []
    for fila in filas:
        valores = fila.split(",")
        registro = dict(zip(cabeceras, valores))
        resultado.append(registro)
    return resultado

# Prueba:
csv_data = [
    "nombre,edad,ciudad",
    "Alice,30,New York",
    "Bob,25,Los Angeles",
    "Charlie,35,Chicago"
]
print("Datos CSV originales:", csv_data)
print("Datos procesados:", procesar_csv(csv_data))
```

### Reto 13: Detector de Palíndromos Absoluto
**Enunciado:** Crea una función que determine si una frase entera es palíndroma (se lee igual de izquierda a derecha que de derecha a izquierda), ignorando espacios y mayúsculas. Ej: "Dabale arroz a la zorra el abad".

**Pistas:** Usa `.replace(" ", "")` y `.lower()`. Luego compara el string resultante con su propia versión invertida usando la sintaxis de corte de listas `[::-1]`.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 13
def es_palindromo(frase):
    limpia = frase.replace(" ", "").lower()
    return limpia == limpia[::-1]

# Prueba:
frase = input("Introduce una frase: ")
if es_palindromo(frase):
    print(f"'{frase}' es un palíndromo.")
else:
    print(f"'{frase}' no es un palíndromo.")
```

### Reto 14: Filtrado de Extensiones
**Enunciado:** Tienes una lista de archivos: `["foto.jpg", "backup_datos.png", "texto.txt", "imagen.png"]`. Crea una función que extraiga solo los que terminen en `.jpg` o `.png`, pero que **no** empiecen por `"backup_"`. Guarda los resultados filtrados como líneas separadas en un archivo `filtrados.txt` (modo append).

**Pistas:** Itera la lista y usa la condición `if arch.endswith((".jpg", \.png")) and not arch.startswith("backup_"):`. Usa `with open()` en modo `"a"` y añade el salto `\n`.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 14
def filtrar_archivos(lista_archivos):
    with open("filtrados.txt", "a") as f:
        for arch in lista_archivos:
            if arch.endswith((".jpg", ".png")) and not arch.startswith("backup_"):
                f.write(arch + "\n")

# Prueba:
archivos = ["foto1.jpg", "prueba.docx", "backup_foto2.png", "documento.pdf", "imagen.png", "backup_imagen.jpg"]
print("Archivos originales:", archivos)
print("Archivos filtrados se guardarán en 'filtrados.txt'")
filtrar_archivos(archivos)
```

### Reto 15: Diccionario Invertido
**Enunciado:** Crea una función que reciba un diccionario cuyas claves y valores sean únicos (ej. `{"a": 1, "b": 2}`) y devuelva un nuevo diccionario donde se hayan intercambiado de posición: los valores ahora son claves, y viceversa.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 15
def invertir_diccionario(dic):
    return {valor: clave for clave, valor in dic.items()}

# Prueba:
diccionario = {"a": 1, "b": 2, "c": 3}
print("Diccionario original:", diccionario)
print("Diccionario invertido:", invertir_diccionario(diccionario))
```

### Reto 16: Boletín de Notas Avanzado
**Enunciado:** Crea una función que reciba una lista irregular como `["Ana", 7, 4, 9, 5]`. El primer elemento siempre es el nombre del alumno y el resto son sus notas. La función debe devolver una tupla con: el nombre, nota máxima, nota mínima y un booleano `True` si aprobó todo (todas las notas >= 5).

**Pistas:** Extrae el nombre con el índice `0` y las notas con segmentación `[1:]`. Usa las funciones `max()`, `min()` y una comprensión de lista envolviendo `all()`.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 16
def analizar_boletin(datos):
    nombre = datos[0]
    notas = datos[1:]
    maxima = max(notas)
    minima = min(notas)
    todo_aprobado = all([n >= 5 for n in notas])
    return (nombre, maxima, minima, todo_aprobado)

# Prueba:
boletin = ["Juan", 7, 8, 6, 9]
resultado = analizar_boletin(boletin)
print("Boletín original:", boletin)
print("Resultado del análisis:", resultado)
boletin2 = ["María", 4, 5, 6, 3]
resultado2 = analizar_boletin(boletin2)
print("Boletín original:", boletin2)
print("Resultado del análisis:", resultado2)
```

### Reto 17: Validador de Coordenadas GPS
**Enunciado:** Crea una función que reciba un texto con formato `"(45.5, -90.2)"`. Debe quitar los paréntesis, separar por la coma, e intentar convertir ambas partes a `float`. Si falla la conversión, devuelve `False`. Si tiene éxito, devuelve una tupla con ambos números.

**Pistas:** Usa `.strip("()")`, `.split(",")`, y un bloque `try/except ValueError` para proteger el Type Casting numérico.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 17
def validar_gps(texto_coord):
    limpio = texto_coord.strip("()")
    partes = limpio.split(",")
    if len(partes) != 2:
        return False
    try:
        lat = float(partes[0])
        lon = float(partes[1])
        return (lat, lon)
    except ValueError:
        return False
    
# Prueba:
coordenada = '(40.7128, -74.0060)'
resultado = validar_gps(coordenada)
if resultado:
    print("Coordenada válida:", resultado)
else:
    print("Coordenada no válida.")

coordenada_invalida = '40,71, -74.0060'  
resultado_invalido = validar_gps(coordenada_invalida)
if resultado_invalido:
    print("Coordenada válida:", resultado_invalido)
else:
    print("Coordenada no válida.")
```

### Reto 18: El Extractor de Iniciales
**Enunciado:** Crea una función que reciba un nombre completo (ej. "juan carlos perez") y devuelva un string solo con sus iniciales en mayúscula y sin puntos (ej. "JCP").

**Pistas:** Separa el string en una lista de palabras con `.split()`. Luego extrae la letra en el índice `[0]` de cada palabra, conviértela con `.upper()`, y guárdalas en una nueva lista o únelas.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 18
def obtener_iniciales(nombre_completo):
    palabras = nombre_completo.split()
    iniciales = [p[0].upper() for p in palabras]
    return "".join(iniciales)

# Prueba:
nombre = input("Introduce un nombre completo: ")
print("Nombre original:", nombre)
print("Iniciales:", obtener_iniciales(nombre))  
```

### Reto 19: Generador de Identificadores
**Enunciado:** Crea una función que reciba el nombre de un usuario y su año de nacimiento (ej. "Ana", 1995). Debe generar y devolver un código único formado por el año, seguido del código numérico ASCII de la primera y última letra del nombre forzadas a mayúscula.

**Pistas:** Usa los índices `[0]` y `[-1]`, el método `.upper()`, y la función `ord()` para obtener el número ASCII correspondiente. Todo debe devolverse como texto continuo.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 19
def generar_id(nombre, anio):
    primera = ord(nombre[0].upper())
    ultima = ord(nombre[-1].upper())
    return f"{anio}{primera}{ultima}"

# Prueba:
nombre = input("Introduce un nombre: ")
anio = input("Introduce un año: ")
id_generado = generar_id(nombre, anio)
print("Nombre original:", nombre)
print("Año:", anio)
print("ID generado:", id_generado)
```

### Reto 20: El Gran Procesador (Reto Final)
**Enunciado:** Escribe un código que pida un nombre de fichero por teclado e intente abrirlo en modo lectura. Si no existe, avisa y el programa termina. Si existe, lee su contenido, conviértelo a minúsculas, elimina los saltos de línea `\n` y extrae a un conjunto (`set`) todas las vocales únicas que aparezcan. Por último, guarda un diccionario con la estructura `{"fichero": nombre, "vocales_usadas": list(vocales)}` en un archivo nuevo llamado `reporte.json`.

**Pistas:** Protege la lectura con `try/except FileNotFoundError`. Limpia con métodos de string. Interseca el texto con un conjunto predefinido de vocales. Importa `json` para volcar los datos.

```python
# Escribe tu código aquí
```

```python
# Solución Reto 20
import json

nombre_archivo = input("Nombre del archivo de texto: ")
try:
    with open(nombre_archivo, "r") as f:
        contenido = f.read()

    limpio = contenido.lower().replace("\n", "")
    vocales = {"a", "e", "i", "o", "u"}
    vocales_encontradas = set(limpio) & vocales

    reporte = {
        "fichero": nombre_archivo,
        "vocales_usadas": list(vocales_encontradas)
    }
 
    with open("reporte.json", "w") as f:
        json.dump(reporte, f)
    print("Reporte generado exitosamente.")

except FileNotFoundError:
    print("El archivo especificado no existe.")
```
