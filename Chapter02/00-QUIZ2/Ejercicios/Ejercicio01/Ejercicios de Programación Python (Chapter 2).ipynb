### Reto 5: Congelando datos (Ficheros JSON)
**Enunciado:** El diccionario del reto anterior se perderá cuando el programa termine. Escribe un código que guarde ese diccionario en un archivo físico llamado `inventario.json` en tu ordenador.

### Reto 7: El tablero de ajedrez (Listas 2D)
**Enunciado:** Crea un tablero de 3x3 (una cuadrícula) relleno de ceros. Luego, simula que colocas una ficha en el centro exacto del tablero cambiando ese cero por un uno. Imprime el tablero fila por fila.

**Pistas:** Una **lista bi-dimensional** es una lista que contiene otras listas: `[[0,0,0], [0,0,0], [0,0,0]]`. Para acceder al centro necesitas doble índice: `tablero[fila][columna]`. Usa un bucle `for fila in tablero:` para imprimir bonito.

```python
# Solución Reto 7
tablero = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
tablero[1][1] = 1  # Centro (índice 1 en fila y columna)

for fila in tablero:
    print(fila)
```

### Reto 8: Clases de distintos tamaños
**Enunciado:** Tienes una lista con los grupos de alumnos de una academia: `[["Ana", "Luis"], ["Pedro"], ["Marta", "Sara", "Juan"]]`. Como ves, no todas las clases tienen los mismos alumnos (lista irregular). Muestra por pantalla cuántos alumnos tiene cada sub-grupo.

**Pistas:** Itera sobre la lista principal con un bucle `for grupo in lista_clases:`. Dentro del bucle, usa la función `len(grupo)` para saber el tamaño de esa sublista concreta.

```python
# Solución Reto 8
lista_clases = [["Ana", "Luis"], ["Pedro"], ["Marta", "Sara", "Juan"]]

for i, grupo in enumerate(lista_clases):
    print(f"El grupo {i+1} tiene {len(grupo)} alumnos.")
```

### Reto 10: Filtrado exprés (Comprehension List)
**Enunciado:** Tienes una lista de números del 1 al 15. Tu jefe te pide que crees una lista nueva *solo* con los números que sean mayores de 10. Tienes prohibido usar más de una línea de código para hacer el filtro.

**Pistas:** Utiliza una **List Comprehension**. La estructura básica es `[variable for variable in lista_original if condicion]`.

```python
# Solución Reto 10
numeros = list(range(1, 16))
mayores_de_10 = [n for n in numeros if n > 10]
print("Lista filtrada:", mayores_de_10)
```

### Reto 11: El contador de letras
**Enunciado:** Tienes el texto "paralelepipedo". Queremos saber exactamente cuántas veces se repite cada letra. Si usas un diccionario normal, te dará error si intentas sumar 1 a una letra que aún no existe. Resuélvelo de la forma más profesional posible.

**Pistas:** Importa la herramienta adecuada: `from collections import defaultdict`. Crea tu contador así: `contador = defaultdict(int)`. Luego haz un bucle `for letra in texto:` y simplemente haz `contador[letra] += 1`.

```python
# Solución Reto 11
from collections import defaultdict

texto = "paralelepipedo"
contador = defaultdict(int)

for letra in texto:
    contador[letra] += 1

print("Frecuencia de letras:", dict(contador))
```

### Reto 15: Control de acceso de seguridad
**Enunciado:** Un sistema tiene 3 sensores en puertas representados en una lista de booleanos: `[True, True, False]`. (True es cerrado, False es abierto). El programa debe avisar si **todas** las puertas están cerradas, y también debe avisar si **alguna** puerta está abierta.

**Pistas:** Usa la función `all(sensores)` para comprobar si todo es True. Para saber si alguna está abierta (es decir, si hay algún False), puedes invertir la lógica temporalmente o buscar si `False in sensores`. También puedes usar `any()` entendiendo cómo evalúa los booleanos.

```python
# Solución Reto 15
sensores = [True, True, False]

if all(sensores):
    print("Todas las puertas están cerradas.")
else:
    print("Atención: no todas las puertas están cerradas.")

if not all(sensores) or False in sensores:
    print("Aviso: Hay alguna puerta abierta.")
```

### Reto 16: Emparejamiento rápido
**Enunciado:** Tienes dos listas separadas. Una con los nombres de tres países `["España", "Francia", "Italia"]` y otra con sus capitales `["Madrid", "París", "Roma"]`. Crea un diccionario que empareje automáticamente cada país con su capital.

**Pistas:** Usa la función `zip(lista1, lista2)` para "cremallerear" ambas listas. Envuelve ese `zip` dentro de la función constructora de diccionarios `dict()`.

```python
# Solución Reto 16
paises = ["España", "Francia", "Italia"]
capitales = ["Madrid", "París", "Roma"]

diccionario_geografico = dict(zip(paises, capitales))
print("Diccionario:", diccionario_geografico)
```

### Reto 20: El desafío integrador final
**Enunciado:** Pide al usuario el nombre de un empleado y su edad. Guarda estos dos datos en un diccionario. Por último, abre un archivo `empleados.txt` en modo añadir (append) y guarda una frase tipo "Nombre: [nombre], Edad: [edad]\n".

**Pistas:** Combina lo aprendido: `input()` para pedir datos, `int()` para convertir la edad, diccionarios `{"nombre": n, "edad": e}` para agrupar en memoria, y `with open(..., "a")` junto con `.write()` para la persistencia.

```python
# Solución Reto 20
nombre = input("Nombre del empleado: ")
edad = int(input("Edad: "))
 
empleado = {"nombre": nombre, "edad": edad}
 
with open("empleados.txt", "a") as f:
    f.write(f"Nombre: {empleado['nombre']}, Edad: {empleado['edad']}\n")
print("Empleado guardado correctamente.")
```
