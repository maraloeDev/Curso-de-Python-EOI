```python
"""Q12.01
Con los datos de tipo tupla, los valores de dos variables se pueden intercambiar sin usar una variable temporal. Usando este método de intercambio, escribe un programa que mueva el valor más grande de la lista dada hacia la última posición.

Guía de codificación:

Usa un bucle for.

Al intercambiar valores en una lista, no se deben usar variables adicionales como temp."""

lst = [5, 6, 3, 9, 2, 12, 3, 8, 7]

for i in range(len(lst) - 1):
    if lst[i] > lst[i + 1]:
        lst[i], lst[i + 1] = lst[i + 1], lst[i]

print(lst)
```

```python
""" Q13.01

Consigna:
El arreglo bidimensional (lista de listas) a contiene los valores [[1, 2], [3, 4], [5, 6]]. Cambia este arreglo bidimensional a uno unidimensional como [1, 2, 3, 4, 5, 6] e imprímelo.

Guía de codificación:

Usa un bucle for.

Define un nuevo arreglo (lista vacía) y coloca los elementos de a dentro de este nuevo arreglo.
"""

lista_original = [[1, 2], [3, 4], [5, 6]]


lista_plana = []

for sublista in lista_original:
    for numero in sublista:

        lista_plana.append(numero)

print(lista_plana)
```

```python
"""Q14.01
Consigna:
Existe una variable de diccionario llamada maria como se muestra abajo. En este diccionario, las materias como 'korean' y 'english' y sus calificaciones se almacenan como clave:valor. Imprime el promedio de las calificaciones de Maria (el resultado debe ser 89.25).

Guía de codificación:

Usa un método de diccionario como .values() y la función len()."""

maria = {'korean': 94, 'english': 91, 'mathematics': 89, 'science': 83}


calificaciones = maria.values()
promedio = sum(calificaciones) / len(calificaciones)

print(promedio)
```
