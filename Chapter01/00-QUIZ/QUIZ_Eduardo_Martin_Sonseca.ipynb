```python
""" Ejercicio 06-01: Sentencia Condicional-1

Escribe un programa que tome dos números enteros aleatorios como entrada y los enumere del más pequeño al más grande.
Condición: Los dos números no son iguales.
Guía: Recibe la entrada del usuario con la función input(), conviértela a entero con int() y usa una sentencia if para comparar y mostrar los valores."""

n1 = str(input("Introduce un numero: "))
n2 = str(input("Introduce un numero: "))

if int(n1) == int(n2):
    print(f"Los numeros son iguales\n Numero 1 = {n1}\n Numero 2 = {n2}")
if n1 < n2:
    print(n1, n2)
else:
    print(n2, n1)
```

```python
"""Ejercicio 07-01: Sentencia Condicional-2

Escribe un programa que realice lo siguiente utilizando una expresión condicional compuesta (anidada):
Guía: El código debe actuar de forma distinta según si la respuesta es 0 o 1.
Ejemplo de flujo:
¿Eres adulto? (1 si lo eres, 0 si eres menor): 1
¿Estás casado? (1 si lo estás, 0 si estás soltero): 1
Resultado: Eres un adulto que está casado.
"""

es_adulto = input("¿Eres adulto? (1: sí, 0: no): ")
if es_adulto == "1":
    está_casado = input("¿Estás casado? (1: sí, 0: no): ")
    if está_casado == "1":
        print("Eres un adulto que está casado.")
```

```python
""" Ejercicio 08-01: Bucles-1

Entre los números naturales positivos distintos de 1, un número que no es primo se llama número compuesto. Imprime los números primos y compuestos del 2 al 12.
Guía: Usa una sentencia for. Si usas un for anidado, la expresión para determinar si es primo debe ir en el bucle interno.
"""

for n in range(2, 13):
    es_primo = True
    for i in range(2, n):
        if n % i == 0:
            es_primo = False
            break
    if es_primo:
        print(f"{n} es primo")
    else:
        print(f"{n} es compuesto")
```

```python
""" Ejercicio 09-01: Bucles-2

Un número de Armstrong es un entero de tres dígitos que es igual a la suma de los cubos de cada uno de sus dígitos. Encuentra todos los números de Armstrong entre los números de tres dígitos (100 al 999) e imprímelos.

Guía: Se deben buscar todos los números del 100 al 999 usando una sentencia for.
"""

for num in range(100, 1000):
    c = num // 100
    d = (num // 10) % 10
    u = num % 10
    if (c**3 + d**3 + u**3) == num:
        print(f"{num} es un número de Armstrong")
```
