```python
"""
Reciba las coordenadas $(x_1, y_1)$ y $(x_2, y_2)$ de dos puntos por parte del usuario e imprima la distancia entre ellos de la siguiente manera: +1Tarea: Implemente la función distance(x1, y1, x2, y2). Guía de codificación: Utilice la siguiente fórmula para calcular la distancia:
"""

import math

def calcular_distancia(x1, y1, x2, y2):

    distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distancia


x1 = float(input("Punto 1 - Coordenada x: "))
y1 = float(input("Punto 1 - Coordenada y: "))
x2 = float(input("Punto 2 - Coordenada x: "))
y2 = float(input("Punto 2 - Coordenada y: "))

print("La distancia es:", calcular_distancia(x1, y1, x2, y2))
```

```python
"""
Pregunta 18-01 Un palíndromo es una oración, palabra o cadena que se lee igual de adelante hacia atrás (ejemplos: level, eve, radar).
+1

Tarea: Use una llamada recursiva para determinar si una palabra es un palíndromo. Defina la función is_palindrome y escriba un programa que reciba una cadena del usuario e imprima si lo es o no.


Guía de codificación: Debe llamar a la función is_palindrome dentro de sí misma.
"""

def es_palindromo(palabra):
    # Si la palabra se queda vacía o con una sola letra, es que todo salió bien
    if len(palabra) <= 1:
        return True

    # Si la primera letra es igual a la última...
    if palabra[0] == palabra[-1]:
        # ...repetimos el proceso pero con lo que queda en el medio
        return es_palindromo(palabra[1:-1])

    return False

texto = input("Escribe una palabra: ")
if es_palindromo(texto.lower()): # .lower() para que no importe si es Mayúscula
    print("¡Sí, es un palíndromo!")
else:
    print("No, no lo es.")
```

```python
"""
Dada una lista llamada n_list con los valores [10, 20, 30]:

Tarea: Escriba un código que pase n_list a la función map y devuelva los siguientes resultados:


Guía de codificación: Use las funciones map y lambda.


Resultados esperados:

El doble del valor de entrada: [20, 40, 60]

El triple del valor de entrada: [30, 60, 90]
"""

numeros = [10, 20, 30]

# Multiplicamos cada número por 2
doble = list(map(lambda x: x * 2, numeros))

# Multiplicamos cada número por 3
triple = list(map(lambda x: x * 3, numeros))

print("Dobles:", doble)
print("Triples:", triple)
```

```python
"""
Analice el siguiente código y determine qué valores se incluirán finalmente en num_list. Calcule el resultado esperado a mano.

def calc_digit(n):
    def final(digit):
        return digit**n
    return final

num_list = []
for num in range(1, 6):
    num_list.append(calc_digit(num))
    # Se imprime el resultado 5 veces
    print(num_list[num-1](num))

Guía de codificación: Prediga la salida que se imprimirá 5 veces a través de la sentencia for.
"""

def fabrica_de_potencias(exponente):
    def calcular(base):
        return base ** exponente
    return calcular

# Creamos una lista vacía para guardar los resultados
resultados = []

for i in range(1, 6): # Del 1 al 5
    # Creamos la función y la usamos al mismo tiempo
    funcion_potencia = fabrica_de_potencias(i)
    valor = funcion_potencia(i)
    resultados.append(valor)
    print(f"Número {i} elevado a {i} es: {valor}")
```

```python
"""
Implemente la multiplicación (*) y la división (/) de dos vectores utilizando los métodos mul y truediv aprendidos en la unidad 21.


Escenario: Suponiendo que v1 es (30, 40) y v2 es (10, 20), escriba un código que imprima la salida como se muestra abajo como resultado de la multiplicación y división de los dos vectores.


Guía de codificación: Escriba un código que extienda la funcionalidad de la clase Vector2D que aprendió en este capítulo.


Salida esperada:

v1 * v2 = (300, 800)

v1 / v2 = (3.0, 2.0)
"""

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self, otro):
        return self.x * otro.x, self.y * otro.y

    def __truediv__(self, otro):
        return self.x / otro.x, self.y / otro.y

v1 = Vector(30, 40)
v2 = Vector(10, 20)

print("Multiplicación:", v1 * v2)
print("División:", v1 / v2)
```
